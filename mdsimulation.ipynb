```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from the text file
data = np.loadtxt('em_potential.xvg')



# Create the plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(data, color='red', linewidth=2, label='Potential Energy')
plt.title('Potential Energy during Minimization')
plt.xlabel('Energy Minimization Step')
plt.ylabel(r'Potential Energy [kJ/mol]')
plt.legend()
plt.grid(True)
plt.show()
```
